## Monte Carlo Simulation

#### Importing relevant libraries

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
Total_topic = 69
Studied_topic = 15
Selected_topic = 5
#Generate a list of 69 topics from 1 to 69
List_of_topics = list(range(1,Total_topic+1))
```

#### Define functions

```python
#Define a function to randomly chose r from n topics
def chose_topics(r):
    return random.sample(List_of_topics, r)
```

```python
#Define a function to find number of common topics between studied and selected
def common_topics(list1, list2):
    return np.intersect1d(list1,list2)
```

```python
#Number of trials to run the simulation
List_of_trials = [100, 1000, 10000, 100000]
#Empty list to record probability for each trial run
Prob_of_success = []
```

#### Execute the simulations and plot the results

```python
#Run a loop for simulating number of trials in the list
for trials in List_of_trials:
    #initialize a counter to count number of success
    success_count = 0
    #loop n number of times the run the experiment
    for i in range(trials):
        #Chosing random topics for exam
        sampled_list = chose_topics(Selected_topic)
        
        #Chosing random topic prepared by student
        prepared_list = chose_topics(Studied_topic)
        
        #Finding the common topic in prepared and chosen topics
        common_in_exam = common_topics(sampled_list, prepared_list)
        
        #Finding the number of common topic in each trial run of the experiment
        #If length of common topics list is larger than zero, this means a cmoomon topic is found and the student is passed
        #Therefore increment the success counter by 1-
        if len(common_in_exam) > 0:
            success_count = success_count+1
    
    #Outside the loop, dividing the number of successes by total experiments in each run
    prob = success_count/trials
    
    #Adding the prob to a list for graphical visualisation
    Prob_of_success.append((prob))

"""""""""""""""""""""""""""""""""""""""""""""""""""""""""
Plotting a log graph of success vs attempts
"""""""""""""""""""""""""""""""""""""""""""""""""""""""""    
plt.style.use('dark_background')
plt.title("Monte Carlo Simulation", fontweight='bold')
plt.ylabel("Probability of Success", fontweight='bold')
plt.xlabel("Number of trials", fontweight='bold')
plt.xscale('log')
plt.grid(True, which="both",  linestyle='--', linewidth=0.6, color = "grey")
plt.ylim(0, 1)

# Plotting the number of runs agains probability of success
plt.plot(List_of_trials,Prob_of_success, color = "orange", marker = "h" )  

#Label the values at each marker point
for x,y in zip(List_of_trials, Prob_of_success):
    label = "{:.4f}".format(y)
    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center',# horizontal alignment 
                 color = 'orange')             

plt.show()
```

### If the topics prepared are 15, then using Monte Carlo simulation , probability of passing is approx 71.48%

### Inference of Monte Carlo Simulation

##### As we run multiple times the number of experiments, it is observed that although in initial experiments when the number of trial runs are low, probability of success has a larger variation. But when the similar experiment is run for 10E4 or 10E5 times, then the variance in probaility is significantly reduced, and almost tends to converge with the value caulcated using theoretical method.
